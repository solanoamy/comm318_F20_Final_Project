## IL LITTLE CAESARS

```python
rchi = requests.get("https://www.menuism.com/restaurant-locations/little-caesars-140160/us/il#chicago")
soupchi = BeautifulSoup(rchi.text, 'html.parser')
```

```python
chicagolinks = []
for chilink in soupchi.find_all('a'):
    found_linkschi = chilink.get('href')
    chicagolinks.append(found_linkschi)
```

```python
chicago_url = chicagolinks[49:189]
```

```python
chicago_pizza_pizza = pd.DataFrame()
```

```python
listtry = []
city = []
zipcode = []
state = []

for url in chicago_url:
    response = requests.get(url)
    newsoup = BeautifulSoup(response.text, "html.parser")
    listtry.append(newsoup.find("span", itemprop="streetAddress").text.strip())
    city.append(newsoup.find("span", itemprop="addressLocality").text.strip())
    zipcode.append(newsoup.find("span", itemprop="postalCode").text.strip())
    state.append(newsoup.find("span", itemprop="addressRegion").text.strip())

    
chicago_pizza_pizza['Street Address'] = listtry
chicago_pizza_pizza['City'] = city
chicago_pizza_pizza['Zip'] = zipcode
chicago_pizza_pizza['State'] = state
```

```python
chicago_pizza_pizza
```

## Doing an address to geo lookup

1. combine the street address, city, state and zip to create a new address column

2. call the gpd.tools.geocode function on this column to get geopoints

3. concatenate the new columns

```python
# add full_address column
phlcaesar['full_address']=phlcaesar[['Street Address','City',
                                     'State', 'Zip']].apply(lambda r: ', '.join(r.values),
                                                            axis=1)
```

```python
phlcaesar
```

```python
# geocode the addresses to get geometry
geoloc=gpd.tools.geocode(phlcaesar['full_address'], provider='nominatim', user_agent="my-application")
```

```python
geoloc
```

```python
geoloc.to_csv('../data/phl_caesar.csv')
```

```python
# combine data frames and turn into a geopandas dataframe

phlcaesar_gdf=gpd.GeoDataFrame(pd.concat([phlcaesar, geoloc], axis=1))
```

```python
phlcaesar_gdf
```

```python
phlcaesar.to_csv('../data/phl_caesar.csv')
```

```python
phlcaesar_gdf.plot()
```
